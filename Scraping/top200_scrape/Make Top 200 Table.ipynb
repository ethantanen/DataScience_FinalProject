```python
import os
import pandas as pd
```

```python
#Get data files in current directory
#After moving this notebook, replace argument with "\dates"
files = os.listdir("data")
```

```python
for a in files:
    print(a)
```

```python
top_tracks = pd.DataFrame(columns=[i.replace(".txt","") for i in files])

top_tracks.head()
```

```python
for i in files:
    print(i)
    tracks = pd.read_csv("data/"+i,sep="\n")
    top_tracks[i.replace(".txt","")] = tracks
    
```

```python
top_tracks.to_csv("top_200_table.csv",sep="\t")
```
