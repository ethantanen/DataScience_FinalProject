```python
dates = []
count = 0

for i in top_100.columns:
    count += 1
    if count%15 == 0:
        dates.append(i)

        
```

```python
for i in range(len(dates)):
    print(i,dates[i])
```

```python
data = pd.DataFrame(index = ['danceability','energy','loudness','tempo','valence'])

for col in dates[249:]:
    
    print(col)
    
    #Data structures to store lists of audio faetures and ids
    audio_features = {'danceability':[],'energy':[],'loudness':[],'tempo':[],'valence':[]}
    ids = {}
    
    #Cycle through year i's tracks and find spotify data on them 
    for track in top_100[col]:
        
        search = spotify.search(track,type="track",limit=1)['tracks']['items']
                
        if(len(search) == 0):
            continue
            
        ids[track] = search[0]['id']
      
    #Get track features using spotipy
    features = spotify.audio_features(ids.values())
    
    #Check if anything was returned 
    if features:
    
        #Update audio features dict w/ respective values 
        for f in features:

            audio_features['danceability'].append(f['danceability'])
            audio_features['energy'].append(f['energy'])
            audio_features['loudness'].append(f['loudness']*-1)
            audio_features['tempo'].append(f['tempo'])
            audio_features['valence'].append(f['valence'])

        #Convert to dataframe 
        df = pd.DataFrame(audio_features)

        #Normalize data
        #df = (df - df.mean()) / (df.max() - df.min())

        #Calculate mean for values for this year
        df = df.mean()
        
        #Add series to chart
        data[col] = df

    




```

```python
data.to_csv('partial_6.csv')
```
