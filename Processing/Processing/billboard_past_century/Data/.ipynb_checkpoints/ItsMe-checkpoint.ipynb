```python
import billboard
import datetime as dt
```

```python
start_date = dt.date(year=1958,month=1,day=1)
day_count = (dt.date.today()-start_date).days
```

```python

top_songs = {}

#Cycle through dates between first available date and todays date
for date in (start_date + dt.timedelta(n) for n in range(day_count)):
    
    #Get chart for date
    chart = billboard.ChartData('hot-100',date=date)
    
    #Create new entry in top_songs dictionary 
    top_songs[date] = [] 
    
    #Cycle through chart
    for song in chart:
        
        #Add song to dict entry w/ date as key
        top_songs[date].append(song.title)
```
