Remove artist and quotation from each entry.

```python
top_100 = pd.read_csv('billboard_top_100_past_decade.csv',index_col=0)

for i in top_100:
    top_100[i] = top_100[i].apply(lambda x: x.split(" by ")[0][1:-1])
```

Create a list of indices. In case the following loop crashes, the loop can be rerun beginning at the last index rather than having to retrieve all the data over again

```python
dates = []
for i in top_100.columns:
    dates.append(i)
```

Cycle through every song and use spotipy to retrieve the features for the song. Average the features for a particular date and add this data to the song_features data frame

```python
top_100.shape
```

```python
for i in range(len(dates)):
    print(i,dates[i])
```

```python
song_features = pd.DataFrame(index = ['danceability','energy','loudness','tempo','valence'])

for col in dates[570:]:
    
    print(col)
    
    #Data structures to store lists of audio faetures and ids
    audio_features = {'danceability':[],'energy':[],'loudness':[],'tempo':[],'valence':[]}
    ids = {}
    
    #Cycle through year i's tracks and find spotify data on them 
    for track in top_100[col]:
        search = spotify.search(track,type="track",limit=1)['tracks']['items']
                
        if(len(search) == 0):
            continue
            
        ids[track] = search[0]['id']
      
    #Get track features using spotipy
    features = spotify.audio_features(ids.values())
    #Check if anything was returned 
    if features:
    
        #Update audio features dict w/ respective values 
        for f in features:
            
            if f == None:
                continue

            audio_features['danceability'].append(f['danceability'])
            audio_features['energy'].append(f['energy'])
            audio_features['loudness'].append(f['loudness']*-1)
            audio_features['tempo'].append(f['tempo'])
            audio_features['valence'].append(f['valence'])

        #Convert to dataframe 
        df = pd.DataFrame(audio_features)

        #Normalize data
        #df = (df - df.mean()) / (df.max() - df.min())

        #Calculate mean for values for this year
        df = df.mean()
        
        #Add series to chart
        song_features[col] = df

    
```

```python
song_features.to_csv("billboard_top_100_past_decade_features_5")
```
