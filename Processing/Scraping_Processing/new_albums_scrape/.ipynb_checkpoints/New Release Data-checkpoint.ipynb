This notebook scrapes the web for upcoming album releases. The notebook begins by importing the necessary modules. The requests module is used to retrieve html from the site with upcoming album information. BeautifulSoup is used for processing the html. And re is used for the data cleaning process.

```python
from bs4 import BeautifulSoup as bs
from requests import get
import re
import pandas as pd
```

The url below 

```python
url = "https://consequenceofsound.net/upcoming-releases/"
```

```python
response = get(url)
#print(response.text[:500])
```

```python
html_soup = bs(response.text, 'html.parser')
```

```python
#<p style="padding-left:30px;">
may_content = html_soup.find('h3',text="May 2018")

html_albums = []

while True:    
    may_content = may_content.findNext('p',style="padding-left:30px;")
    
    if may_content.em:
        html_albums.append(may_content)
    
    if may_content.findNext('h3').text == "August 2018":
        del html_albums[-1]
        break

```

```python
artists_albums = []

for a in html_albums:
    
    artist = a.strong.text.replace("\xa0","").lstrip()
    artist = re.sub(r'\([^)]*\)', '', artist).replace(")","").rstrip()
    
    album = a.em.text.replace("\xa0","").lstrip()
    album = re.sub(r'\([^)]*\)',"",album).replace(")","").rstrip()
    
    artists_albums.append([artist,album])
   
```

```python
dicts = {'artists':[],'albums':[]}

for a in artists_albums:
    dicts['artists'].append(a[0])
    dicts['albums'].append(a[1])
    
dicts
```

```python
new_albums = pd.DataFrame(dicts)
```

```python
new_albums.to_csv("new_albums.csv")
```
