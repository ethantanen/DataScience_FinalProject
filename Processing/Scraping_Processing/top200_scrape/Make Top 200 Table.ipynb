This notebook takes the top 200 songs for the last year and a half or so and creates a table from the data. First, the necessary modules must be imported. The os module is used to list the names of files in the directory that contains the top 200 charts. 

```python
import os
import pandas as pd
```

The following line of code creates a list of file names. The files are kept in the directory "data", which can be view outside of this notebook.

```python
files = os.listdir("data")
```

Next, an empty data frame is created. The column names are simply the dates for which we have data. The .txt extension is removed.

```python
top_tracks = pd.DataFrame(columns=[i.replace(".txt","") for i in files])
```

Now the program iterates over the file names, creates a series from the iformation and adds the data to the top_tracks data frame.

```python
for i in files:
    tracks = pd.read_csv("data/"+i,sep="\n")
    top_tracks[i.replace(".txt","")] = tracks
    
```

It's always wise to check whether or not the data looks right. Generally, the head will be printed off but we'd also like to see data on the shape of the dataframe, which can be found at the bottom of the printed frame. The data frame passes the sniff test; there are 200 rows, which aligns with the expectation of 200 of the top songs per day. 

```python
top_tracks
```

Not every day has usable data. The files that have bad data begin with an "<html>" tag while the usable data begins with a songs position on the chart. To remove the bad files we simply check to see if the first piece of information is a number and delete the columns for which the first piece of information is "<html>", more specifically not a number. The isdigit function is perfect for the job!

```python
for i in top_tracks:
    if not top_tracks[i][0].split(",")[0].isdigit():
        del top_tracks[i]
```

The spotify charts website includes a bunch of metadata on each track, which we dont need for the analysis. The following code removes all the metadata save for the artists name, the tracks name, the songs ranking and the number of views.

```python
for j in top_tracks:
    top_tracks[j] = [(i.split(",")[0] + "," + i.split(",")[1] + "," + i.split(",")[2] + "," + i.split(",")[3]) for i in top_tracks[j]]
      
```
