# Secondary Datasets and Graphs

From the intial, raw spotify data, another chart was created that tracks a songs lifespan in the top 200. The rows of this new chart are dates in which a parituclar song is either on or off the charts and there is a column per song. The NaN value is used to denote a song not being on the chart.

```python
song_tracker = pd.read_csv("song_tracker.csv",low_memory=False,na_values="NaN")
del song_tracker["Unnamed: 0"]
del song_tracker["infinity (888) - feat. Joey Bada$$"]

song_tracker = song_tracker[1:]
song_tracker = song_tracker.set_index(top_200.columns)
song_tracker.index = pd.to_datetime(song_tracker.index)
song_tracker.head()
```

The table below shows the lifespan of Post Malone's song "Rockstar". The songs position on the chart is denoted by the y-axis and the day of the year is denoted by the x-axis. The chart is a little confusing in that the lower the value the higher the song was on the chart. Immedietly following is a chart for the Migo's song "Bad and Boujee (feat. Lil Uzi Vert)"

```python
plot = song_tracker["rockstar"][1:].astype(float).plot(figsize=(20,10),title="Rockstar by Post Malone")
plot.set_xlabel("Date")
plot.set_ylabel("Chart Position")
```

```python
plot = song_tracker["Bad and Boujee (feat. Lil Uzi Vert)"][1:].astype(float).plot(figsize=(20,10),title="Bad and Boujee by Migos")
plot.set_xlabel("Date")
plot.set_ylabel("Chart Position")
```

It's clear that the lifespan of a popular song is quite long with a predictable decline in ranking as time goes on. Here are the two charts superimposed on each other. It gives you a little perspective into the misleadingness of charts!


```python
plot = song_tracker[["rockstar","Bad and Boujee (feat. Lil Uzi Vert)"]].astype(float).plot(figsize=(20,10))
plot.set_xlabel("Date")
plot.set_ylabel("Chart Position")
```

Further pre-processing can be done on the data to see how often an artist apears in the top 200 tracks per month. This work is illustrated in a notebook titled "Make Song Count Table". Basically, for a particular month, the number of times an artists has a song in the top 200 is recorded and added to a dictionary, which is later converted to a dataframe and csv file. 

We now import the csv file produced in the aforementioned notebook. A quick plot shows that the artist Cardi B dominated the charts in April. 

```python
df = pd.read_csv("song_count.csv",index_col=0)
plt.bar(range(len(df.Count)),df.Count,align="center")
plt.xticks(range(len(df.Count)),df.index,rotation=70)
plt.figure(figsize=(20,10))
```
