```python
import spotipy
import spotipy.util as util
```

```python
#Get Token
token = util.prompt_for_user_token(username="ethantanen")
```

```python
#Get spotify object
sp = spotipy.Spotify(auth=token)
```

```python
#Get song features
query="Bob Dylan"

results = sp.search(q=query,type='artist',limit=10)
artists_id = ""

for artists in results['artists']['items']:
    if artists['name'] == query:
        artists_id = artists['id']
        break
```

```python
#features = sp.audio_features(tids)
```

```python
#Get artists top albums 
albums = sp.artist_albums(artists_id, album_type=None, country=None, limit=20, offset=0)
```

```python
#Get album id's
albums_ids = []

for a in albums['items']:
   albums_ids.append(a['id'])
```

```python
#Get album objects for tracks 
tracks = []

for i in albums_ids:
     tracks.extend(sp.album_tracks(i)['items'])
```

```python
#Look at track names 
for i in tracks:
    print("")
    print(sp.track(i['id'])['name'])
```
